### test in monocyte

```python
def robust_cov(mod, cluster, params, fsc=1):
    # compute bread
    bread = np.linalg.inv(mod.hessian(params))
    
    # compute meat
    score_obs = mod.score_obs(params) 
    ind_by_cluster = pd.get_dummies(cluster).values
    score_cluster = score_obs.T @ ind_by_cluster
    meat = score_cluster @ score_cluster.T
    
    return fsc * (bread @ meat @ bread)    
```

```python
celltype_var = 'cluster'
cluster_var = 'Donor'

adata_cd14 = adata[adata.obs[celltype_var]=='CD14 Mono',:].copy()
adata_cd14.obs.loc[:,'random_treat'] = np.random.binomial(n=1, p=0.5, size=adata_cd14.shape[0])
gene_nz = np.asarray(adata_cd14.raw.X.sum(axis=0) > 10).ravel()
Y = adata_cd14.raw.X[:,gene_nz]
X = sm.add_constant(pd.get_dummies(adata_cd14.obs['random_treat'], drop_first=True)).values
cluster = adata_cd14.obs[cluster_var]
sf = np.asarray(adata_cd14.raw.X.sum(axis=1)).ravel()
sf = sf / sf.mean()
```

```python
%%time
bse_HC0 = np.zeros(Y.shape[1])
bse_cluster= np.zeros(Y.shape[1])

p_HC0 = np.zeros(Y.shape[1])
p_nb = np.zeros(Y.shape[1])
for gidx in range(Y.shape[1]):
    mod = sm.GLM(endog=Y[:,gidx].toarray().ravel(), exog=X, family=sm.families.Poisson(), offset=np.log(sf))
    fit = mod.fit(cov_type='HC0')
    fit2 = sm.GLM(endog=Y[:,gidx].toarray().ravel(), exog=X, family=sm.families.NegativeBinomial(), offset=np.log(sf)).fit()
    bse_HC0[gidx] = fit.bse[1]
    #bse_cluster[gidx] = np.sqrt(np.diag(robust_cov(mod, cluster, fit.params, fsc=6/5))[1])
    
    p_HC0[gidx] = fit.pvalues[1]
    p_nb[gidx] = fit2.pvalues[1]
```

```python
%%time
bse_HC0 = np.zeros(Y.shape[1])
bse_cluster= np.zeros(Y.shape[1])

p_HC0 = np.zeros(Y.shape[1])
p_nb = np.zeros(Y.shape[1])
for gidx in range(Y.shape[1]):
    mod = sm.GLM(endog=Y[:,gidx].toarray().ravel(), exog=X, family=sm.families.Poisson(), offset=np.log(sf))
    fit = mod.fit(cov_type='HC0', method='newton')
    #fit2 = sm.GLM(endog=Y[:,gidx].toarray().ravel(), exog=X, family=sm.families.NegativeBinomial(), offset=np.log(sf)).fit()
    bse_HC0[gidx] = fit.bse[1]
    #bse_cluster[gidx] = np.sqrt(np.diag(robust_cov(mod, cluster, fit.params, fsc=6/5))[1])
    
    p_HC0[gidx] = fit.pvalues[1]
    p_nb[gidx] = fit2.pvalues[1]
```

## No pseudobulk - Fixest

```python
%%time
%%R -i Y -i X -i sf -o pval_fixest
pval_fixest <- c()
for (i in 1:ncol(Y)){
    y <- as.integer(Y[,i])
    df <- as.data.frame(cbind(y,X,sf))
    fit <- fixest::fepois(y~V3, offset=as.numeric(sf), data=df, vcov='HC')
    pval_fixest <- c(pval_fixest,fixest::pvalue(fit)[2])
}
```

```python
%%R 
dds <- DESeq2::DESeqDataSetFromMatrix(Yt,
                        colData = X,
                        design = ~ 1+V2)
sizeFactors(dds) <- as.numeric(sf)
dds <- DESeq2::estimateDispersions(dds, quiet = TRUE)
dds <- DESeq2::nbinomWaldTest(dds, minmu = 1e-6)
```

## No pseudbulk - MAST

```python
%%R
Xdf <- as.data.frame(X)
Xdf$cluster <- cluster
sce <- SingleCellExperiment(list(counts=Yt), colData=Xdf)
```

```python
%%R
sce <- logNormCounts(sce)
```

```python
%%R
sca <- SceToSingleCellAssay(sce)
```

```python
%%R 
fit_mast <- zlm(formula = ~V2+(1|cluster), sca=sca, method='glmer', ebayes = FALSE, parallel=TRUE)
```

```python
%%R
test_mast <- lrTest(fit_mast, 'V2')
```

```python
%%R -o pval_mast
pval_mast <- (test_mast[,,'Pr(>Chisq)'][,3])
```
